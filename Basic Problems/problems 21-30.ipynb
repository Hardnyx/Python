Write a Python program that determines whether a given number (accepted from the user) is even or odd, and prints an appropriate message to the user.

```python
def odd_even_checker(x:int):
    if not isinstance(x,int):
        raise Exception('Non valid number')
    mod = x & 2
    if mod == 1:
        print('The number is odd')
    else:
        print('The number is even')

odd_even_checker(2)
```

Write a Python program to count the number 4 in a given list.

```python
def four_counter(l:list):
    return l.count(4)

four_counter([1,2,3,4,4,4])

# Proposed way is to define a count value, iterate for each number and increase the value
# each time the number is equal to four. Works fine for other programming languages
```

Write a Python program to get n (non-negative integer) copies of the first 2 characters of a given string. Return n copies of the whole string if the length is less than 2.

```python
def two_first_characters(text:str, x:int):
    if not isinstance(x, int):
        raise Exception('Non valid number')
    if len(text) == 1:
        return text*x
    else:
        return text[:2] * x
    
two_first_characters("Hello", 3)
```

Write a Python program to test whether a passed letter is a vowel or not.

```python
def vowel_test(letter:str):
    if not isinstance(letter, str):
        raise Exception('Non valid value')
    if len(letter) > 1:
        print('You should pass only a char')
    else:
        letter = letter.upper()
        vowels = ['A','E','I','O','U']
        if letter in vowels:
            print('The letter is a vowel')
        else:
            print('The letter is not a vowel')
        
vowel_test('B')
```
