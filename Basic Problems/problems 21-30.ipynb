Write a Python program that determines whether a given number (accepted from the user) is even or odd, and prints an appropriate message to the user.

```python
def odd_even_checker(x:int):
    if not isinstance(x,int):
        raise Exception('Non valid number')
    mod = x & 2
    if mod == 1:
        print('The number is odd')
    else:
        print('The number is even')

odd_even_checker(2)
```

Write a Python program to get n (non-negative integer) copies of the first 2 characters of a given string. Return n copies of the whole string if the length is less than 2.

```python
def two_first_characters(text:str, x:int):
    if not isinstance(x, int):
        raise Exception('Non valid number')
    if len(text) == 1:
        return text*x
    else:
        return text[:2] * x
    
two_first_characters("Hello", 3)
```

Write a Python program to test whether a passed letter is a vowel or not.

```python
def vowel_test(letter:str):
    if not isinstance(letter, str):
        raise Exception('Non valid value')
    if len(letter) > 1:
        print('You should pass only a char')
    else:
        letter = letter.upper()
        vowels = ['A','E','I','O','U']
        if letter in vowels:
            print('The letter is a vowel')
        else:
            print('The letter is not a vowel')
        
vowel_test('B')
```

Write a Python program that checks whether a specified value is contained within a group of values.
Test Data :
3 -> [1, 5, 8, 3] : True
-1 -> [1, 5, 8, 3] : False

```python
def value_in_collection(value, col):
    col_type = type(col).__name__
    match col_type:
        case 'list':
            print('The inserted collection is a list')
        case 'tuple':
            print('The inserted collection is a tuple')
        case 'set':
            print('The inserted collection is a set')
        case _:
            raise Exception('Not a collection')
        
    return value in col

value_in_collection('A',(1,2,'A'))
```

Write a Python program to create a histogram from a given list of integers.


```python
# As the specified solution

def histogram_integer_counter(l:list):
    for x in l:
        if not isinstance(x, int):
            raise Exception('Not an integer')
        print('*'*x)

histogram_integer_counter([1,4,3,2])
```

```python
# As I understood

def histogram_of_integers(l:list):
    l_set = set(l)
    for x in l_set:
        if not isinstance(x, int):
            raise Exception('Not an integer')
        h = '*'*l.count(x)
        print(f'{x}: {h}')

histogram_of_integers([2,2,2,3,4,5,5,5,5])
```

Write a Python program that concatenates all elements in a list into a string and returns it.

```python
def list_concatenator(l:list):
    text = ""
    for x in range(len(l)):
        text += l[x]
    return text

list_concatenator(['A','B','C'])
```

Write a Python program that prints out all colors from color_list_1 that are not present in color_list_2.
Test Data :
color_list_1 = set(["White", "Black", "Red"])
color_list_2 = set(["Red", "Green"])
Expected Output :
{'Black', 'White'}

```python
def print_list_1_excluding_list_2(color_list_1, color_list_2):
    
    if not isinstance(color_list_1, (list, set)) or not isinstance(color_list_2, (list, set)):
        raise Exception('Non valid argument')
    if type(color_list_1).__name__ == 'list': color_list_1 = set(color_list_1)
    if type(color_list_2).__name__ == 'list': color_list_2 = set(color_list_2)

    color_list_difference = color_list_1.difference(color_list_2)

    for color in color_list_difference:
        print(color)

print_list_1_excluding_list_2(["White", "Black", "Red"],["Red", "Green"])
```

Write a Python program that will accept the base and height of a triangle and compute its area.


```python
def triangle_area(base:float, height:float):
    if not isinstance(base, (float, int)) or not isinstance(height, (float, int)): raise Exception('Non valid type')
    if base <= 0 or height <= 0: raise Exception('Non valid value')
    area = base * height / 2
    return(area)

triangle_area(2,3)
```
