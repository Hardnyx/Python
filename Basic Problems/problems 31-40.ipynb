Write a Python program that computes the greatest common divisor (GCD) of two positive integers.


```python
def GCD_calculator(x1:int,x2:int):

    if not isinstance(x1, int) or not isinstance(x2, int): raise TypeError("Arguments must be integers")
    if x1 <= 0 or x2 <= 0: raise ValueError("Arguments must be positive integers")

    def divisors_calculator(x:int):
        
        if x % 2 == 0:
            u_limit = int(x/2)+1 
        else:
            u_limit = int((x+1)/2)

        divisors = []

        divisors = [j for j in range(1, u_limit) if x % j == 0]
        divisors.append(x)

        print(divisors)
        return divisors

    divisors_x1 = divisors_calculator(x1)
    divisors_x2 = divisors_calculator(x2)

    common_divisors = sorted(list(set(divisors_x1).intersection(divisors_x2)))
    print(common_divisors)

    GCD = max(common_divisors)
    print(f'The GCD is {GCD}')
    return GCD

GCD_calculator(10**7, 10**7-1)
```

Write a Python program to find the least common multiple (LCM) of two positive integers.

```python
import math as m
def LCM_calculator(*args):

    values = list(args)
    if any(not isinstance(x, int) for x in values): raise TypeError("Arguments must be integers")
    if any(x < 0 for x in values): raise ValueError("Arguments must be positive integers")

    u_limit = m.ceil(max(values)/2)

    LCM = 1

    for d in range(2, u_limit + 1):
        print(f'Divisor number {d}')
        while any(x > 1 for x in values):
            while any(x % d == 0 for x in values):
                if any(x % d == 0 for x in values):
                    LCM = LCM * d
                    for i in range(len(values)):
                        if values[i] % d == 0:
                            print(f'The number {values[i]} is divisible by {d}. The result is {int(values[i] / d)}')
                            values[i] = int(values[i] / d)
                        else:
                            print(f'The number {values[i]} is not divisible by {d}')
                    print(LCM)
                    print(f'Next pass for {d}')


            print(values)
            
            if d >= 3:
                d = d+2
                print(f'Ended with {d}. Continuing with {d+2}')
            else:
                d = d+1
                print(f'Ended with {d}. Continuing with {d+1}')
    
    print(f'The LCM is {LCM}')
    return(LCM)

LCM_calculator(4,5,36)
```

Write a Python program to sum three given integers. However, if two values are equal, the sum will be zero.

```python
def special_sum_for_three_integers(x1:int, x2:int, x3:int):
    
        if any(not isinstance(x, int) for x in (x1,x2)): raise TypeError("Arguments must be integers")

        if x1 == x2 or x1 == x3 or x2 == x3:
            return 0
        else:
            return x1 + x2 + x3
        
special_sum_for_three_integers(2,3,3)
```

Write a Python program to sum two given integers. However, if the sum is between 15 and 20 it will return 20.


```python
def special_sum_ceil_to_20(x1:int, x2:int) -> int:

    if any(not isinstance(x, int) for x in (x1,x2)): raise TypeError("Arguments must be integers")

    if x1 + x2 in range(15,20):
        return 20
    else:
        return x1 + x2
    
special_sum_ceil_to_20(10,5)
```

Write a Python program that returns true if the two given integer values are equal or their sum or difference is 5.

```python
def equal_sum_difference_5(x1: int, x2: int) -> bool:

    if any(not isinstance(x, int) for x in (x1,x2)): raise TypeError("Arguments must be integers")

    if x1 == x2: return True
    elif abs(x1 - x2) == 5: return True
    elif x1 + x2 == 5: return True
    else: return False


equal_sum_difference_5(4,3)
```

Write a Python program to solve (x + y) * (x + y).
Test Data : x = 4, y = 3
Expected Output : (4 + 3) ^ 2) = 49

```python
def square_number(x:float, y:float) -> float:
    if any(not isinstance(x, (int, float)) for x in (x,y)): raise TypeError("Arguments must be numeric values")
    return (x+y)**2

square_number(1,3)
```

Write a Python program to compute the future value of a specified principal amount, rate of interest, and number of years.
Test Data : amt = 10000, int = 3.5, years = 7
Expected Output : 12722.79

```python
def compound_value(amt:int, i:float, years:int) -> int:
    if any(not isinstance(x, (int, float)) for x in (amt, i, years)): raise TypeError("Arguments must be numeric values")
    return round(amt*(1+i/100)**years,2)

compound_value(10000,3.5,7)
```

Write a Python program to calculate the distance between the points (x1, y1) and (x2, y2).

```python
import math as m

def vectorial_norm(x:tuple, y:tuple)-> float:

    if any(not isinstance(t, tuple) for t in (x, y)): raise TypeError("Arguments must be tuples")
    if len(x) != len(y): raise TypeError("Arguments must have the same length")

    return m.sqrt(sum((a - b)**2 for a, b in zip(x, y)))

vectorial_norm((1,1),(4,5))
```
