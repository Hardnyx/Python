Write a Python program that calculates the area of a circle based on the radius entered by the user.
Sample Output :
r = 1.1
Area = 3.8013271108436504

```python
import math as m
def circle_area():
    r = float(input('Enter the radius of the circle: '))
    area = m.pi * r ** 2
    print(f"The circle's area is {area}")
circle_area()
```

Write a Python program that accepts the user's first and last name and prints them in reverse order with a space between them.

```python
def name_inverter():
    first_name = input('Insert your first name: ')
    second_name = input('Insert your second name: ')
    name_inverted = second_name + ' ' + first_name
    print(f'Your name inverted is {name_inverted}')

name_inverter()
```

Write a Python program that accepts a sequence of comma-separated numbers from the user and generates a list and a tuple of those numbers.

```python
def tuple_generator():
    sequence = input('Insert a sequence of comma separated nombers')
    numbers_tuple = tuple(sequence.split(','))
    print(f'The tuple {numbers_tuple} was generated')

tuple_generator()
```

Write a Python program that accepts a filename from the user and prints the extension of the file.

```python
def filename_spliter():
    filename = input('Insert the filename: ')
    if '.' not in filename or filename.endswith('.'):
        return 'The name given has no extension'
    name, extension = filename.rsplit('.',1)
    print (f"The name of the file is {name} and it's a {extension} file")

filename_spliter()
```

Write a Python program to display the examination schedule. (extract the date from exam_st_date).

Write a Python program that accepts an integer (n) and computes the value of n+nn+nnn.
Sample value of n is 5
Expected Result : 615

```python
def n_transformation():
    n = float(input('Insert an integer: '))
    if n%1 != 0:
        return "The number isn't an integer"
    n = int(n)
    if n >= 0:
        nn = int(str(n)*2)
        nnn = int(str(n)*3)
        number = n + nn + nnn
        print(f'The number is {number}')
    else:
        return 'The number is negative'

n_transformation()
```
