Sample function : abs()
Expected Result :
abs(number) -> number

```python
def documentation_printer(function:str):
    print(function.__doc__)
documentation_printer(abs)
```

Write a Python program that prints the calendar for a given month and year.
Note : Use 'calendar' module.

```python
import calendar
def calendar_printer():
    while True:
        try:
            y = int(input('Insert the year: '))
            if y <= 0:
                print('The year is not positive')
                continue
            break
        except ValueError as v:
            return f'Type error: {v}'       
    while True:
        try:
            m = int(input('Insert the number of the month: '))
            if m < 1 or m > 12:
                print('The month is out of range')
                continue
            print(calendar.month(y,m))
            break
        except ValueError as v:
            return f'Type error: {v}'
        except Exception as e:
            return f'Unhandled exception: {e}'
        

calendar_printer()
```
