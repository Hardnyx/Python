Write a Python program that prints the calendar for a given month and year.
Note : Use 'calendar' module.

```python
import calendar
def calendar_printer():
    while True:
        try:
            y = int(input('Insert the year: '))
            if y <= 0:
                print('The year is not positive')
                continue
            break
        except ValueError as v:
            return f'Type error: {v}'       
    while True:
        try:
            m = int(input('Insert the number of the month: '))
            if m < 1 or m > 12:
                print('The month is out of range')
                continue
            print(calendar.month(y,m))
            break
        except ValueError as v:
            return f'Type error: {v}'
        except Exception as e:
            return f'Unhandled exception: {e}'
        

calendar_printer()
```

Write a Python program to calculate the number of days between two dates.
Sample dates : (2014, 7, 2), (2014, 7, 11)
Expected output : 9 days

```python
import datetime
def days_difference():

    # First

    while True:
        date1 = input('Insert the first year, month and day, separated by commas: ')
        if date1.count(',') != 2:
            print('Insert the date in the given format')
            continue
        try:
            y1, m1, d1 = [int(x) for x in tuple(date1.split(','))]
        except ValueError:
            print(ValueError('Insert the data properly'))
            continue
        break

    # Second

    while True:
        date2 = input('Insert the second year, month and day, separated by commas: ')
        if date2.count(',') != 2:
            print('Insert the date in the given format')
            continue
        try:
            y2, m2, d2 = [int(x) for x in tuple(date2.split(','))]
        except ValueError:
            print(ValueError('Insert the data properly'))
            continue
        break    

    date1 = datetime.date(y1, m1, d1)
    date2 = datetime.date(y2, m2, d2)
    
    diff = abs(date1 - date2).days
    print(f"The days difference is {diff}")

days_difference()
```

Write a Python program to get the volume of a sphere with radius six.


```python
import math as m
def sphere_volume():
    while True:
        try:
            r = float(input('Insert the radius of the sphere: '))
            if r <= 0:
                raise ValueError
            break
        except ValueError:
            print('Insert a positive number')
            continue

    volume = 4/3 * m.pi * r**3
    print(f'The sphere volume is {volume}')
        
sphere_volume()
```
